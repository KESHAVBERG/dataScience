```
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df = pd.read_excel('train.xlsx')
```

```
df.head()
```

How salary and job are related

```
df_ = df.head(300)
```

```
Tail_df_ = df.tail(300)
```

```
df_.drop('Unnamed: 0' , axis=1,inplace = True)
```

which board gives highest paid jobs

```
fig = plt.gcf()
fig.set_size_inches(25,6)

sns.barplot(y='Salary' ,x = '12board', data=Tail_df_)
plt.xticks(rotation=90);
```

```
fig = plt.gcf()
fig.set_size_inches(15,6)

sns.barplot(y='Salary' ,x = '12board', data=df_)
plt.xticks(rotation=90);
```
